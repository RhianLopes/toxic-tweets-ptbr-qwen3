# 02b — Preparar Dataset de Avaliação Completo

Gera `data/full/toldBr_eval.csv` a partir do dataset raw, removendo duplicatas.

- Entrada: `data/raw/toldBr_full.csv` (21.000 tweets)
- Saída: `data/full/toldBr_eval.csv` (~20.639 tweets após deduplicação)

```python
from pathlib import Path
import polars as pl

Path("../data/full").mkdir(parents=True, exist_ok=True)

df_raw = pl.read_csv("../data/raw/toldBr_full.csv")
print(f"Raw: {len(df_raw)} tweets | colunas: {df_raw.columns}")
```

```python
# Deduplicação por texto
duplicados = len(df_raw) - df_raw.n_unique(subset=["text"])
print(f"Duplicatas encontradas: {duplicados}")

df_eval = df_raw.unique(subset=["text"], keep="first", maintain_order=True)
print(f"Após deduplicação: {len(df_eval)} tweets")
```

```python
# Distribuição de labels
print(df_eval["label"].value_counts().sort("count", descending=True))
```

```python
df_eval.write_csv("../data/full/toldBr_eval.csv")
print(f"Salvo: data/full/toldBr_eval.csv ({len(df_eval)} tweets)")
```
